```python
import matplotlib.pyplot as plt
import numpy as np

# Simple data to display in various forms
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)
```

```python
# for a long list of plots, like showing a list of images
fig, axes = plt.subplots(2, 2, figsize=(10, 10))
for ax in axes.flatten():
    ax.plot(x, y)
plt.show()
```

```python
# when there's a known number of plots
plt.subplots(2, 2, figsize=(10, 10))
plt.subplot(2, 2, 1)
plt.plot(x, y, c='r')
plt.subplot(2, 2, 2)
plt.plot(x, y, c='g')
plt.subplot(2, 2, 3)
plt.plot(x, y, c='b')
plt.subplot(2, 2, 4)
plt.plot(x, y, c='cyan')
```
